# 00 — Prerequisites

Before running the workshop notebooks, complete the two setup steps below.

---

## Step 1 — Create a Unity Catalog Volume

1. Open **Catalog Explorer** in your Databricks workspace
2. Navigate to your catalog and schema (create them if needed)
3. Click **Create Volume** and name it `raw_data`

Your volume path will be:
```
/Volumes/{catalog}/{schema}/raw_data/
```

---

## Step 2 — Upload the 3 CSV Files

Upload the following files into volume:

| File | Rows | Description |
|---|---|---|
| `account_targeting.csv` | 300 | Account-level targeting with penetration metrics |
| `hcp_procedure_volume.csv` | 150 | Healthcare provider procedure volumes by NPI |
| `product_upgrades.csv` | 400 | Product upgrade opportunities by account and territory |

**How to upload:**
1. In Catalog Explorer, open your volume
2. Click **Upload to this volume**
3. Upload all 3 CSV files into that subfolder

Final paths should look like:
```
/Volumes/{catalog}/{schema}/raw_data/account_targeting.csv
/Volumes/{catalog}/{schema}/raw_data/hcp_procedure_volume.csv
/Volumes/{catalog}/{schema}/raw_data/product_upgrades.csv
```

---

## Next Steps
